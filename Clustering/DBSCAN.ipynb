### Working of DBSCAN
1. Core Points: A point is a core point if it has at least a minimum number of points (minPts) within a specified radius (eps), forming a cluster.
2. Border Points: A point that has fewer than minPts neighbors but is in the neighborhood of a core point is a border point.
3. Noise Points: Points that are not core points and are too far away from core points are considered noise or outliers.

### Parameters:
1. eps (epsilon): The maximum distance between two points to consider them neighbors.(Radius of a specific circle)
2. minPts: The minimum number of points required to form a dense region (cluster).

```python
from sklearn.cluster import DBSCAN
import numpy as np

# Sample data
X = np.array([[1,2], [2,2], [2,3], [8,7], [8,8], [25,80]])

# DBSCAN model
dbscan = DBSCAN(eps=3, min_samples=2)

# Fit model
dbscan.fit(X)

# Cluster labels
labels = dbscan.labels_
print(labels)
```

### why to use dbscan?
1. To skip the outliers while clustering we can use dbscan

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN, KMeans
```

```python
# Generate a sample dataset
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=0)
```

```python
# Apply DBSCAN
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan.fit_predict(X)
```

```python
# Apply KMeans
kmeans = KMeans(n_clusters=4)
kmeans_labels = kmeans.fit_predict(X)
```

```python
# Plotting the clusters
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12,6))

# DBSCAN Plot
ax1.scatter(X[:,0], X[:,1], c=dbscan_labels, cmap='plasma')
ax1.set_title("DBSCAN Clustering")
ax1.set_xlabel("Feature 1")
ax1.set_ylabel("Feature 2")

# KMeans Plot
ax2.scatter(X[:,0], X[:,1], c=kmeans_labels, cmap='plasma')
ax2.set_title("KMeans Clustering")
ax2.set_xlabel("Feature 1")
ax2.set_ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
