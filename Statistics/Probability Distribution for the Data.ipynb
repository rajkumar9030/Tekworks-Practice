## Bernoulli Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import bernoulli

# Define probability of success
p = 0.6

# Generate 1000 Bernoulli samples
data = bernoulli.rvs(p, size=1000)

# Plot the distribution
plt.bar([0, 1], [sum(data == 0), sum(data == 1)], color=['red', 'green'], tick_label=["Failure (0)", "Success (1)"])
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Benoulli Distribution (p=0.6)")
plt.show()
```

## Binomial Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Define parameters
n = 10
p = 0.3

# Generate Binomial distribution data
x = np.arange(0, n+1)
pmf = binom.pmf(x, n, p)

# Plot the Binomial Distribution
plt.bar(x, pmf, color='blue', alpha=0.7)
plt.xlabel("Number of Successes (Clicks)")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")
plt.show()
```

## Poisson Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Define Lambda (average number of events per interval)
lam = 3

# Generate Poission Distribution Data
x = np.arange(0, 10) # Number of events(calls per minute)
pmf = poisson.pmf(x, lam)

# Plot the Poission Distribution
plt.bar(x, pmf, color='purple', alpha=0.7)
plt.xlabel("Number of Calls per Minute")
plt.ylabel("Probability")
plt.title(f"Poisson Distribution (lambda={lam})")
plt.show()
```

## Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define parameters
mu = 70 # Mean
sigma = 10 # Standard Deviation

# Generate Normal Distribution Data
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100) #Range of scores
pdf = norm.pdf(x, mu, sigma) #Compute PDF

# Plot the Normal Distribution
plt.plot(x, pdf, color='red', label=f"N({mu}, {sigma}^2)")
plt.xlabel("Exam Scores")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (mu={mu}, sigma={sigma})")
plt.legend()
plt.grid()
plt.show()
```

## Standard Normal Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate Standard Normal Distribution Data
z = np.linspace(-4, 4, 1000) #Z-values from -4 to 4
pdf = norm.pdf(z, 0, 1) # Standard Normal Distribution (mu=0, sigma=1)

# Plot the Standard Normal Distribution
plt.plot(z, pdf, color='blue', label="Standard Normal Distribution (mu=0, sigma=1)")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.title("Standard Normal Distribution (Z-Distribution)")
plt.axvline(x=0, color='red', linestyle='--', label="Mean (Z=0)")
plt.legend()
plt.grid()
plt.show()
```

## Central Limit Theorem

```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

# Set random seed for reproducibility 
np.random.seed(42)

# Generate a skewed income dataset (exponential distribution) 
income_data = np.random.exponential (scale=50, size=1000)

# Plot the original skewed distribution
plt.figure(figsize=(8, 4))
sns.histplot(income_data, bins=50, kde=True, color="r") 
plt.title("Original Income Distribution (Skewed)") 
plt.xlabel("Income ($1000)")
plt.ylabel("Frequency")
plt.show()

# Perform sampling and calculate means
sample_means = []
n_samples = 1000 # Number of samples
sample_size = 30 # Sample size


for _ in range(n_samples):
    sample = np.random.choice(income_data, size=sample_size) 
    sample_means.append(np.mean(sample))
    
# Plot the sampling distribution of the sample mean
plt.figure(figsize=(8, 4))
sns.histplot(sample_means, bins=50, kde=True, color="b")
plt.title(f"Sampling Distribution of Sample Mean (n= {sample_size})")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.show()
```
