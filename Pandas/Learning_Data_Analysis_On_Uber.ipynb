```python
# Add a column MILES_CAT and based on MILES categorize as Long or Short trip
import numpy as np
uber_data["MILES_CAT"] = np.where(uber_data["MILES*"] > 100, "Long trip", "Short trip")
uber_data.head()
```

```python
# Add another column nc and assigning value 10 will adds the value for all the fields
uber_data["nc"] = 10
```

```python
# Add a column MILES_CATEGORY and based on MILES categorize as Long or Short or Medium trip
# if miles <= 100 -- Short trip
# miles b/w 100 to 200 -- Medium trip
# miles>200 -- Long trip
uber_data["MILES_CATEGORY"] = np.where(uber_data["MILES*"] <= 100, "Short trip", np.where(uber_data["MILES*"] <= 200, "Medium trip", "Long trip"))
```

```python
uber_data
```

```python
# Drop the column MILES_CAT
uber_data.drop(columns="MILES_CAT")
```

```python
# # Add a column MILES_CAT and based on MILES categorize as Long or Short or Medium trip
conditions = [(uber_data["MILES*"] > 100), (uber_data["MILES*"] > 50) & (uber_data["MILES*"] <= 100), (uber_data["MILES*"] <=50)]
categories = ["Long trip", "Medium trip", "Short trip"]
uber_data["MILES_CAT"] = np.select(conditions, categories, default="Unknown")
uber_data
```

```python
# Find the mean of MILES based on the PURPOSE
uber_data.groupby("PURPOSE*")["MILES*"].agg("mean")
```

```python
# Find average miles of each purpose
grouped = uber_data.groupby("CATEGORY*")["MILES*"].agg(["sum","mean","max"])
grouped
```
