```python
def r2score(y_obs,y_pred):
    
    yhat = np.mean(y_obs)
    
    ss_res = 0
    ss_tot = 0
    
    for i in range(len(y_obs)):
        ss_tot += (y_obs[i] - yhat) ** 2
        ss_res += (y_obs[i] - y_pred[i]) ** 2
        
    r2 = 1 - (ss_res / ss_tot)

    return r2
```
