```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y)

plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x, y, color='blue', linestyle='dashed', linewidth=2, marker='o')

plt.title("Customized Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [1000,4000,2000,6000,3000,5000]

plt.plot(x, y, color='green', linestyle='dashed', marker='o')

plt.title("Month vs Savings amount Line Plot")
plt.xlabel("Month")
plt.ylabel("Savings amount")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]

plt.bar(categories, values, color='skyblue')

plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]

plt.bar(categories, values, color=['red', 'blue', 'green', 'violet', 'orange'])

plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]

plt.bar(categories, values, color=['red', 'blue', 'green', 'violet', 'orange'], edgecolor='black')

plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
categories = ['A', 'B', 'C', 'D', 'E']
values = [10, 15, 7, 12, 9]

plt.bar(categories, values, color=['red', 'blue', 'green', 'violet', 'orange'], edgecolor='black', width=0.5)

plt.title("Basic Bar Plot")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
ages = [20, 22, 22, 23, 25, 27, 28, 29, 30, 31, 32, 34, 35, 35, 37, 38, 38, 40, 42, 43, 45, 45, 47, 50, 52, 53, 53, 54, 57, 58, 60, 61, 62, 62]
b = [20, 30, 40, 50, 60, 70]

plt.hist(ages, bins=b, color='skyblue', edgecolor='black')

plt.title("Age Distribution of customers")
plt.xlabel("Age Ranges")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
labels = ['Python', 'Java', 'C++', 'Ruby']
sizes = [40, 30, 20, 10] 
colors=['gold', 'lightblue', 'lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.legend()
plt.title("Programming Language Usage")
plt.show()
```

```python
x = [5, 10, 15, 20, 25, 30, 35]
y = [10, 20, 25, 30, 35, 40, 45]

plt.scatter(x, y, color='green', s=100, edgecolor='black')

plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```
