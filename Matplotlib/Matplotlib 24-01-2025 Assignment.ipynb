```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('sales_data.csv')
df1 = pd.read_csv('customer_data.csv')
df2 = pd.read_csv('school_performance_data.csv')
```

```python
#line plot to track monthly sales
df['Month_Only'] = pd.to_datetime(df['Month']).dt.strftime('%b')
plt.plot(df['Month_Only'],df['Total_Sales'],color = 'orange')
plt.grid(True)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Histogram for Order Quantity
b = 10
plt.hist(df['Order_Quantity'], bins= b, color='orange', edgecolor='black')
plt.grid(True)
plt.title('Order Quantity Distribution')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()
```

```python
#Customer Age Distribution
c = 30
plt.hist(df1['Age'], bins= c, color='lightblue', edgecolor='black')
plt.grid(True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
#Scatter Plot
plt.scatter(df1['Age'],df1['Spending_Amount'],color ='maroon',edgecolors='gold')
plt.grid(True)
plt.title('Age vs Spending Amount')
plt.xlabel('Age')
plt.ylabel('Spending Amount')
plt.xticks(range(0,100,10))
plt.yticks(range(0,1000,100))
plt.show()
```

```python
#bar Plot
subset = df2.iloc[:3]
x = subset['Student_ID']
plt.bar(x, subset['Maths'], label='Maths', width=0.3)
plt.bar(x, subset['Science'], label='Science', width=0.3,bottom=subset['Maths'])
plt.bar(x, subset['English'], label='English', width=0.3,bottom=subset['Maths'] + subset['Science'])
plt.legend()
plt.title('Marks Comparison for Students')
plt.xlabel('Students')
plt.ylabel('Marks')
plt.show()
```

```python
# Box plot
plt.boxplot([df2['Maths'], df2['Science'], df2['English']], labels=['Maths', 'Science', 'English'])
plt.title("Marks Distribution Across Subjects")
plt.xlabel("Subjects")
plt.ylabel("Marks")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
