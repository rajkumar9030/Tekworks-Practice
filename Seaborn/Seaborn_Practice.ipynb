```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("train.csv")
data.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(data=data, x="Pclass", y="Survived", errorbar=None)
plt.title("Survival Rate by Passenger Class", fontsize=16)
plt.xlabel("Passenger Class", fontsize=12)
plt.ylabel("Survival Rate", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(data=data, x="Age", y="Fare", hue="Survived")
plt.title("Relationship between Age and Fare", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Fare", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(12,6))
subset_columns = ["Age", "Fare", "Pclass", "Survived"]
sns.pairplot(data=data[subset_columns], hue="Survived", diag_kind="kde")
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

```python
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
plt.figure(figsize=(12,6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```
